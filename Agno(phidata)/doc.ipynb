```python
# create vectordb
vector_db = Qdrant(
    collection="phidata-qdrant-ytipynb",
    url=QDRANT_URL,
    api_key=QDRANT_API_KEY,
    embedder = OllamaEmbedder()
)
```

```python
URL = "https://www.morganstanley.com/content/dam/msdotcom/en/assets/pdfs/Morgan_Stanley_2023_ESG_Report.pdf"

knowledge_base = PDFUrlKnowledgeBase(
    urls = [URL],
    vector_db = vector_db,
)

# Load the knowledge base: Comment after first run as the knowledge base is already loaded
knowledge_base.load(upsert=True)
```

```python
Agent??
```

```python
agent = Agent(
    model=Groq(id="llama3-70b-8192"),
    #model=OpenAIChat(id="gpt-4o"),
    knowledge=knowledge_base,
    # Enable RAG by adding context from the `knowledge` to the user prompt.
    add_context = True,
    # Set as False because Agents default to `search_knowledge=True`
    search_knowledge=False,
    show_tool_calls=True,
    markdown=True,
    
    storage=SqlAgentStorage(table_name="phidata", db_file="agents_rag.db"),
    add_history_to_messages=True,
)

agent.print_response("How do we achive Sustainable Finance goal", stream=True)
```

```python

from phi.model.azure import AzureOpenAIChat
load_dotenv()
azure_model = AzureOpenAIChat(
    id=os.getenv("AZURE_OPENAI_MODEL_NAME"),
    api_key=os.getenv("AZURE_OPENAI_API_KEY"),
    azure_endpoint=os.getenv("AZURE_OPENAI_ENDPOINT"),
    azure_deployment=os.getenv("AZURE_OPENAI_DEPLOYMENT"),
)
agent_openai = Agent(
    #model=Groq(id="llama3-70b-8192", api_key=GROQ_API_KEY),
    model=azure_model,
    knowledge=knowledge_base,
    # Set as False because Agents default to `search_knowledge=True`
    search_knowledge=False,
    show_tool_calls=True,
    markdown=True,
    storage=SqlAgentStorage(table_name="phidata", db_file="agents_rag.db"),
    add_history_to_messages=True,
)

agent_openai.print_response(
  "what is goal of this 2023 ESG Report document ", stream=True
)
```

```python

agent_ollama = Agent(
    #model=Groq(id="llama3-70b-8192"),
    #model=OpenAIChat(id="gpt-4o"),
    model=Ollama(id="llama3.2:latest"),
    knowledge=knowledge_base,
    # Set as False because Agents default to `search_knowledge=True`
    search_knowledge=False,
    show_tool_calls=True,
    markdown=True,
    
    storage=SqlAgentStorage(table_name="phidata", db_file="agents_rag.db"),
    add_history_to_messages=True,
)

agent_ollama.print_response(
  "what is goal of this 2023 ESG Report document", stream=True
)
```

Query sqlite

```python
import sqlite3

# Connect to the SQLite database
conn = sqlite3.connect('agents_rag.db')

# Create a cursor object
cursor = conn.cursor()

# Execute a query to retrieve all records from the 'phidata' table
cursor.execute('SELECT * FROM phidata')

# Fetch all rows from the executed query
rows = cursor.fetchall()

# Process and print the results
for row in rows:
    print(row)

# Close the cursor and connection
cursor.close()
conn.close()
```
